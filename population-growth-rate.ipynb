```python
N = xG(t) * FG(xA(t)) + (1 - xG(t)) * Fg(xA(t))
```

```python
V = sym.diff(N, t).doit() / N
```

$$ \frac{\dot{N}(t)}{N(t)} = \frac{\bigg(x_G(t)\frac{\partial F_G(x_A(t))}{\partial x_A(t)} + (1 - x_G(t))\frac{\partial F_g(x_A(t))}{\partial x_A(t)}\bigg)\dot{x}_A(t) + \bigg(F_G(x_A(t)) - F_g(x_A(t))\bigg)\dot{x}_G(t)}{x_G(t)F_G(x_A(t)) + (1 - x_G(t))F_g(x_A(t))} $$

which can be re-written as 

$$ \frac{\dot{N}(t)}{N(t)} = \left(\frac{1}{N(t)}\frac{\partial N(t)}{\partial x_A(t)}\right)\dot{x}_A(t) + \left(\frac{1}{N(t)}\frac{\partial N(t)}{\partial x_G(t)}\right)\dot{x}_G(t) $$


Suppose that we define 

$$ V(t) = \frac{\dot{N}(t)}{N(t)} $$

then 

$$ \dot{V}(t) = \frac{\ddot{N}(t)}{N(t)} - \left(\frac{\dot{N}(t)}{N(t)}\right)^2 $$

and 

\begin{align}
    \frac{\partial V(t)}{\partial x_A(t)} = \frac{1}{N(t)}\frac{\partial \dot{N}(t)}{\partial x_A(t)} - \frac{\dot{N}(t)}{N(t)}\frac{1}{N(t)}\frac{\partial N(t)}{\partial x_A(t)}
\end{align}

$$ \ddot{N}(t) = \bigg(x_G(t)\frac{\partial F_G(x_A(t))}{\partial x_A(t)} + (1 - x_G(t))\frac{\partial F_g(x_A(t))}{\partial x_A(t)}\bigg)\ddot{x}_A(t) + \bigg(x_G(t)\frac{\partial^2 F_G(x_A(t))}{\partial x_A(t)^2} + (1 - x_G(t))\frac{\partial^2 F_g(x_A(t))}{\partial x_A(t)^2}\bigg)\dot{x}_A(t)^2 + \bigg(F_G(x_A(t)) - F_g(x_A(t))\bigg)\ddot{x}_G(t) + 2\bigg(\frac{\partial F_G(x_A(t))}{\partial x_A(t)} - \frac{\partial F_g(x_A(t))}{\partial x_A(t)}\bigg)\dot{x}_A(t)\dot{x}_G(t) $$

which can be written as 

$$ \ddot{N}(t) = \frac{\partial N(t)}{\partial x_A(t)}\ddot{x}_A(t) + \frac{\partial^2 N(t)}{\partial x_A(t)^2}\dot{x}_A(t)^2 + \frac{\partial N(t)}{\partial x_G(t)}\ddot{x}_G(t) + 2\frac{\partial^2 N(t)}{\partial x_G(t)\partial x_A(t)}\dot{x}_A(t)\dot{x}_G(t) $$
